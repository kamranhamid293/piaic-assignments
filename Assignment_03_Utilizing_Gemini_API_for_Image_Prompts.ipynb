```python
import google.generativeai as genai
genai.configure(api_key=GEMINI_API_KEY)
```

```python
from google.generativeai.generative_models import GenerativeModel
model: GenerativeModel = genai.GenerativeModel("gemini-1.5-flash")
```

#Image Prompting

```python
!curl -o image.jpg "https://cdn.prod.website-files.com/61a05ff14c09ecacc06eec05/66e8522cbe3d357b8434826a_ai-agents-p-2000.jpg"
```

```python
import PIL.Image
img = PIL.Image.open('image.jpg')
display(img)
```

```python
prompt = """Explain the image"""
```

```python
response = model.generate_content([prompt, img])
print(response.text)
```
